```python
#loading the data
import pandas as pd
laqn = pd.read_csv("laqn_7days.csv")
```

```python
#cleaning data to only NO2 data from one sensor
NO2 = laqn[(laqn['site_code'] == 'EA6') & (laqn['species_code'] == ("NO2"))]
NO2 = NO2.sort_values(by='measurement_end_utc')
NO2.shape
NO2.to_csv("NO4data.csv")
```

```python
#plotting data
NO2["measurement_end_utc"] = pd.to_datetime(NO2["measurement_end_utc"])
NO2.plot("measurement_end_utc", "value")
```

```python
# Training data
X_train, y_train = NO2.measurement_end_utc[:166], NO2.value[:166]

# Test data
X_test, y_test = NO2.measurement_end_utc[166:], NO2.value[166:]
```
